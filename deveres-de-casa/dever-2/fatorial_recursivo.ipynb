# Calcula o fatorial de um número utilizando recursão.

Parâmetro:
     n (int): número inteiro não negativo

Retorno:
     int: fatorial de n

```python
def fatorial(n):
    # fatorial de 0 ou 1 é 1
    if n <= 1:
        return 1

    # chamada recursiva
    return n * fatorial(n - 1)
```

teste simples da função:

```python
numero = int(input("Digite um número inteiro: "))

resultado = fatorial(numero)

print(f"Fatorial de {numero} = {resultado}")
```

Mede o tempo de execução do cálculo de fatorial para diferentes valores de n.

```python
# Nefinindo valores para n
valores = [50, 100, 500, 1000]
tempos_execucao = []

print(f"{'n':<5} | {'Tempo (s)':<12} | {'Dígitos':<8} | {'Resultado Completo'}")
print("-" * 80)

for n in valores:
    inicio = time.perf_counter()
    resultado = fatorial(n)
    fim  = time.perf_counter()
        
    tempo_execucao = fim - inicio
    tempos_execucao.append(tempo_execucao)
    
    #número de dígitos do resultado
    digitos = len(str(resultado))

    # print(f"{n:<5} | {tempo_execucao:<12.6f} | {digitos:<8} | {resultado}")
    print(f"{n:<5} | {tempo_execucao:<12.6f} | {digitos:<8} ")
```

Gráfico:

Eixo x = valores de n

Eixo y: tempo de execução

```python
plt.figure(figsize=(8, 5))

plt.plot(valores, tempos_execucao, marker='o')

plt.title("Tempo de execução do fatorial recursivo")
plt.xlabel("Valor de n")
plt.ylabel("Tempo de execução (segundos)")

plt.grid(True)

plt.show()
```

# O algoritmo faz n chamadas recursivas, assim, o crescimento do tempo tende a ser linear.
# Isso confirma a complexidade: T(n) = O(n)
