# Este Notebook foi criado para comparar empiricamente a eficiência de dois métodos de ordenação: Insertion Sort O(n^2) e a função sorted() que usa Timsort O(n log n).

------------------------------------------------------------------------------------------------------------

Bibliotecas

```python
import time
import random
```

insertion_sort

```python
def insertion_sort(lista):
    for i in range(1,len(lista)):
        chave = lista[i]
        j =  i - 1
        while j >= 0 and lista[j] > chave:
            lista[j + 1] = lista[j]
            j -= 1
        lista[j + 1] = chave
```

Gerar lista aleatória e fazer testes

```python
# Configuraçôes
tamanhos = [1000, 5000, 10000, 20000, 50000]
tempo_ins = []
tempo_tim = []

print(f"{'n':<10} | {'Insertion (s)':<15} | {'Sorted() (s)':<15}")
print("-" * 50)
for n in tamanhos:
    # Gerando lista aleatória que servirá de base
    lista_original = [random.randint(0, n) for _ in range(n)]

    # Teste Insertion Sort
    lista_ins = lista_original.copy()
    inicio = time.perf_counter()
    insertion_sort(lista_ins)
    fim = time.perf_counter()
    t_calculado_ins = fim - inicio
    tempo_ins.append(t_calculado_ins)

    # Teste Timsort (sorted())
    lista_tim = lista_original.copy()
    inicio = time.perf_counter()
    sorted(lista_tim)
    fim = time.perf_counter()
    t_calculado_tim = fim - inicio
    tempo_tim.append(t_calculado_tim)

    print(f"{n:<10d} | {t_calculado_ins:<15.5f} | {t_calculado_tim:<15.5f}")
```
